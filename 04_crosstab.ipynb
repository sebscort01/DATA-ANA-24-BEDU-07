```python
import pandas as pd
import squarify
import matplotlib.pyplot as plt

# Cargar el dataset
cybersecurity_data = pd.read_csv('tmp/Ejemplo_02_03_Cybersecurity_Dataset.csv') 
cybersecurity_data.head()
```

```python
# Crear una tabla de informacion cruzada (num de incidentes por severidad y tipo de ataque)
cross_data = pd.crosstab(cybersecurity_data['severity'], cybersecurity_data['attack_type'])
cross_data
```

```python
cross_data.plot(
    kind='bar',
    stacked=True,
    figsize=(10,6),
    colormap='viridis'
)

plt.title('num de incidentes por severidad y tipo de ataque', fontsize=16)
plt.xlabel('Severidad del incidente', fontsize=14)
plt.ylabel('Numero de incidentes', fontsize=14)

plt.legend(title='attack_type', bbox_to_anchor=(1, 1), loc='upper left')

plt.show()
```
