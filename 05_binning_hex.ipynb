```python
import pandas as pd
import squarify
import matplotlib.pyplot as plt
import seaborn as sns

# Cargar el dataset
cybersecurity_data = pd.read_csv('tmp/Ejemplo_02_03_Cybersecurity_Dataset.csv') 
cybersecurity_data.head()
```

```python
# Crear jointplot
sns.jointplot(
    x='severity',
    y='response_time',
    data=cybersecurity_data,
    kind='hex',
    cmap='Blues',
    height=8
)

plt.suptitle('Binning Hex', y=1.02, fontsize=16)

plt.show()
```
